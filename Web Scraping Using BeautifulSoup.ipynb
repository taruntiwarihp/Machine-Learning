```python
import requests
from bs4 import BeautifulSoup
```

```python
response=requests.get("https://www.flipkart.com/search?q=mobiles&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_1_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_1_0_na_na_na&as-pos=1&as-type=TRENDING&suggestionId=mobiles&requestId=a7f33b27-b93c-4bf8-8b9a-a934ec3b4fe1")
print(response)#2xx=>Success
```

```python
soup=BeautifulSoup(response.content,"html.parser")
print(soup)
```

```python
title_div=soup.find("div",attrs={"class":"_3wU53n"})#Get 
print(title_div.text)
```

```python
title_divs=soup.find_all("div",attrs={"class":"_3wU53n"})
title=[title.text for title in title_divs]
print(title)
```
