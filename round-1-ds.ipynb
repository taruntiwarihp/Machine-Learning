#### Q4 What all operations can you perform on string before applying tfidf vectorizer (theory).

Perform following operations on string
1. lowercase<br>
2. remove digits and punct<br>
3. remove stopwords<br> 

```python
st = "My Name is Tarun Tiwari, my ID no 1234, #$%"
```

```python
 val = st.lower()
```

```python
import re
val = " ".join(re.findall('[a-zA-Z]+',val))
```

```python
from nltk.corpus import stopwords
words = set(stopwords.words('english'))
```

```python
from nltk.tokenize import word_tokenize
word_tokens = word_tokenize(val)
filtered_sentence = [w for w in word_tokens if not w.lower() in words]
 
filtered_sentence = []
 
for w in word_tokens:
    if w not in words:
        filtered_sentence.append(w)
```
